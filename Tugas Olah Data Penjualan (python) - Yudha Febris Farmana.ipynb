Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

memasukkan dataset

```python
df = pd.read_csv("supermarket_sales.csv")
```

Tugas pertama, lakukan cleaning dataset terlebih dahulu untuk mendeteksi dan memperbaiki (atau menghapus) suatu record yang 'corrupt' atau tidak akurat!

```python
# Identifikasi missing value
df.dropna(inplace=True)

# Identifikasi duplikat
df.drop_duplicates(inplace=True)

df.to_csv("data_penjualan_bersih.csv", index=False)

print(df)
```

Tugas kedua, dari dataset penjualan yang telah disajikan, gunakan fungsi dasar pemrograman Python untuk mencari data di bawah ini!
a.	Data branch yang memiliki rating paling tinggi
b.	Data product line yang memiliki gross income paling besar


```python
# Mencari data dengan rating tertinggi
branch_max_rating = df[df['Rating'] == df['Rating'].max()]
print("Data dengan rating tertinggi:")
print(branch_max_rating)

# Mencari data dengan gross income tertinggi
product_line_max_income = df[df['gross income'] == df['gross income'].max()]
print("Data dengan gross income tertinggi:")
print(product_line_max_income)
```

Tugas ketiga, buatlah pie chart yang membandingkan banyaknya pembeli laki-laki dan perempuan!

```python
# Membuat pie chart untuk membandingkan jumlah pembeli laki-laki dan perempuan
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Pembeli Laki-laki vs Pembeli Perempuan')
plt.show()
```

Tugas keempat, hitunglah revenue penjualan secara keseluruhan. Lalu buatlah bar chart yang berisi total revenue tersebut!

```python
# Menghitung revenue
df['revenue'] = df['Unit price'] * df['Quantity']
total_revenue = df['revenue'].sum()

# Membuat plot bar
plt.figure(figsize=(8, 6))
plt.bar('Total Revenue', total_revenue, color='skyblue', edgecolor='black')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.title('Total Revenue Penjualan')

# Menambahkan label pada bar
plt.text('Total Revenue', total_revenue, f'Total: {total_revenue:,.2f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

# Menampilkan grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
